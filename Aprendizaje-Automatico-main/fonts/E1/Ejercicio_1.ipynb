```julia
using DelimitedFiles

function separador_dir()
    if Sys.iswindows()
        return "\\"
    else
        return "/"
    end
end
sep = separador_dir()


dataset = readdlm("iris_database$sep"*"iris.data", ',')

inputs = dataset[:, 1:4]
targets = dataset[:, 5]

println(size(inputs))
println(size(targets))

function normalize(x, classes)
    if length(classes) == 2
       x == classes[0] ? true : false
    else
        return classes .== x
    end
end

function normalizeCategorics(targets)
    classes = unique(targets)
    vector = normalize.(targets, Ref(classes))
    return hcat(vector...)'
end

inputs = convert(Array{Float32, 2}, inputs)
targets = normalizeCategorics(targets)
println(targets)
println(typeof(targets))

@assert (size(inputs,1) == size(targets,1)) "Las matrices de entradas y salidas deseadas no tienen el mismo número de filas"
```

```julia
using DelimitedFiles

function separador_dir()
    if Sys.iswindows()
        return "\\"
    else
        return "/"
    end
end
sep = separador_dir()


dataset = readdlm("iris_database$sep"*"iris.data", ',')

inputs = dataset[:, 1:4]
targets = dataset[:, 5]

println(size(inputs))
println(size(targets))

function normalize(targets)
    classes = unique(targets)
    if length(classes) == 2
        two_clasees(x) = x == classes[1] ? true : false
        encoded_values = two_clasees.(targets)
    else
        onehoting(x) = classes .== x
        encoded_values = hcat(onehoting.(targets)...)'
    end
    return encoded_values
end

inputs = convert(Array{Float32, 2}, inputs)
targets = normalize(targets)
println(targets)
println(typeof(targets))

@assert (size(inputs,1) == size(targets,1)) "Las matrices de entradas y salidas deseadas no tienen el mismo número de filas"
```

```julia
using Statistics

#Extraemos los datos de máximo, mínimo, media, desviación típica de cada columna (atributo)
#El resultado será una matriz de una fila y tantas columnas como atributos

min_values = minimum(inputs, dims=1)
max_values = maximum(inputs, dims=1)
mean_values = mean(inputs, dims=1)
std_values = std(inputs, dims=1)

function normalizeMaxMin(i)
    n_atributes = size(i,2)
    #Recorremos los atributos de inputs
    #Para cada atributo  
    for atribute in 1:n_atributes
        #si el máximo es igual al mínimo, le asignamos un valor de 0 a toda la columna
        if max_values[atribute] == min_values[atribute]
            i[:, atribute] .= 0
        #sino aplicamos la normalización 
        else 
            i[:, atribute] = (i[:, atribute] .- min_values[atribute]) ./ (max_values[atribute] .- min_values[atribute])
        end
    end
    return i
end

#Sin bucles
function normalizeMaxMin2(i)
    i .= ifelse.(max_values .== min_values, 0, (i .- min_values) ./ (max_values .- min_values))
    return i
end


#println(normalizeMaxMin(inputs))
println(normalizeMaxMin2(inputs))
println(size(inputs))

function normalizeMeanStd(i)
    n_atributes = size(i,2)
    #Recorremos los atributos de inputs
    #Para cada atributo  
    for atribute in 1:n_atributes
        #si la desviación típica es igual a 0, le asignamos un valor de 0 a toda la columna
        if std_values[atribute] == 0
            i[:, atribute] .= 0
        #sino aplicamos la normalización 
        else 
            i[:, atribute] =  (i[:, atribute] .- mean_values[atribute]) ./ std_values[atribute]
        end
    end
    return i
end

#println(normalizeMeanStd(inputs))

```
